```python
import numpy as np
import matplotlib.pyplot as plt    
```

```python

N = 11 #number of grid points
L = 1 #domain size
h = np.float64(L/(N-1)) #grid spacing

iterations = 0 #number of iterations

#Initializing temperature field
T = np.zeros(N) #temperature array
T[N-1] = 1 #boundary condition at x=L

#Inititalizing iterated temperature
T_new = np.zeros(N)
T_new[N-1] = 1 

#Error related parameters
epsilon = 1e-8
numerical_error = 1.0 #initial error

#checking the error tolerance
while numerical_error > epsilon:
    #Computing for all interior points
    for i in range(1,N-1):
        T_new[i] = 0.5*(T[i-1] + T[i+1])

    #resetting the numerical error and recalculate
    numerical_error = 0.0
    for i in range(1,N-1):
        numerical_error += np.abs( T[i] - T_new[i] )

    #iteration advancement
    iterations += 1
    T = T_new.copy()

#Plotting the results
x_dom = np.arange(N)*h

plt.plot(x_dom, T, 'ro-',linewidth=2, markersize=10)
plt.xlabel('x',size=20)
plt.ylabel('T',size=20)
plt.title('Temperature distribution in a 1D rod')
plt.grid(True, color = 'k')
plt.show()

```
